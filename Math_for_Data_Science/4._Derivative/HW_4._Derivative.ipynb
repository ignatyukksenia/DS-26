### Уровень 0:
Посчитайте значение первой и второй производной функции:

$f(x) = x^5 + 4sin(2x) + cos(3x+3)$

В точке x = 1.

```python
import numpy as np
from sympy import symbols, sin, cos, diff, evalf
from scipy.misc import derivative
import matplotlib.pyplot as plt
```

```python
# Создание символьной переменной
x = symbols('x')

# Описание функции
func = x**5+4**sin(2*x)+cos(3*x+3)

# Вычисление первой и второй производных функции
first_der = diff(func, x, 1).evalf(4, subs = {x: 1})
second_der = diff(func, x, 2).evalf(4, subs = {x: 1})
```

```python
print('Значение первой производной при x = 1:', first_der)
print('Значение второй производной при x = 1:', second_der)
```

### Уровень 1:¶
Постройте график первой и второй производной функции:

$f(x) = sin^5(2x+1)$

На отрезке [-5; 5].

```python
# Задание функции
def f(X):
    return np.sin(X**2+1)
```

```python
# Определение интервала для вычисления производной
X = np.linspace(-5, 5)

# Вычисление значений функции на отрезке (-5; 5)
fx = f(X)

# Вычисление певой и второй производной
df1 = [derivative(f, xx, dx = 1e-6, n = 1) for xx in X]
df2 = [derivative(f, xx, dx = 1e-6, n = 2) for xx in X]

# Построение графиков первой и второй производной
plt.figure(figsize = (14, 8))
plt.plot(X, fx, label = 'function')
plt.plot(X, df1, label = '1st derivative')
plt.plot(X, df2, label = '2nd derivative')
plt.legend()
plt.show()
```
