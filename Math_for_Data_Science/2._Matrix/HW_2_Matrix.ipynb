### Уровень 0:
* Создайте матрицу 3 на 3, заполненную числами от 1 до 9
* Прибавьте к ней матрицу с единицами на диагонале
* Умножьте полученную матрицу на вектор (1, 2, 3)
* Посчитайте обратную матрицу к полученной матрице

```python
import numpy as np
from numpy import linalg
import matplotlib.pyplot as plt
```

```python
# Создание матрицы 3x3, заполненной числами от 1 до 9

a = np.reshape(range(1, 10), (3, 3))
a
```

```python
# Прибавление единичной матрицы 3х3

b = a + np.eye(3)
b
```

```python
# Умножение полученной матрицы на вектор (1, 2, 3)

c = b @ np.array([1, 2, 3])
c
```

У полученной матрицы нет обратной, т.к. она не является квардатной.

### Уровень 1:

Дан объект в 2d пространстве. 

При помощи матричных операций выполните:

* Уменьшить объект в два раза
* Повернуть на 130 градусов
* Отразить объект относительно прямой y=x
* Выполнить смещение объекта и масштабирование (применить 2 матрицы трансформаций). Смещение на (200, 300) и масштабирование на (0.5, 1.2)

```python
# задание объекта в 2D пространстве

A = [
[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]
]

A = np.array(A)
x = A[:,0]
y = A[:,1]

plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
# Уменьшение объекта в два раза

e = np.array([x, y])

f = 0.5 * e

plt.plot(f[0], f[1])
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
# Поворот объекта на 130°

cos = np.cos(130/ 180 * np.pi)
sin = np.sin(130/ 180 * np.pi)
a = A @ np.array([[cos, -sin, 0], [sin, cos, 0], [0, 0, 1]])

x_1 = a[:,0]
y_1 = a[:,1]

plt.figure(figsize = (8, 8))
plt.plot(x_1, y_1)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
# Отображение объекта относительно оси y=x

inversed = A @ np.array([[0, 1, 0], [1, 0, 0], [0, 0, 1]])


x_2 = inversed[:,0]
y_2 = inversed[:,1]

plt.plot(x_2, y_2)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
#  Смещение объекта на (200, 300) и масштабирование на (0.5, 1.2)

x_3 = (x + 200) * 0.5
y_3 = (y + 300) * 1.2

plt.plot(x_3, y_3)
plt.ylim([-200, 600])
plt.xlim([-200, 600])
```
