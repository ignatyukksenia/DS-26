### Задание 1. Написать функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг

Результат классификации записать в столбец class

```python
import pandas as pd
```

```python
def ranging(param):
    if param >= 4.5:
        return 'high'
    elif 2 < param <= 4:
        return 'medium'
    else:
        return 'low'
```

```python
df = pd.read_csv('ml-latest-small/ratings.csv')
total_rating = df.groupby('movieId').mean()                          # Подсчет рейтинга по каждому фильму
total_rating['class'] = total_rating['rating'].apply(ranging)        # Классификация рейтингов
total_rating.sort_values(by = 'rating', ascending = False).head(20)  # Сортировка по убыванию рейтинга
```

### Задание 2.
Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определенному региону. Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим ‘undefined’.

```python
geo_data = {'Центр': ['москва', 'тула', 'ярославль'],
            'Северо-Запад': ['петербург', 'псков', 'мурманск'],
            'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']}
```

```python
def geo_classification(param):
    for k, v in geo_data.items():
        for el in v:
            if el in param:
                return k
    return 'undefined'
```

```python
file = pd.read_csv('ml-latest-small/keywords.csv')

# Добавление столбца 'region' с указанием гео-метки
file['region'] = file['keyword'].apply(geo_classification)
file.sample(20)
```

### Задание 3 (бонусное)

Есть мнение, что “раньше снимали настоящее кино, не то что сейчас”. Ваша задача проверить это утверждение, используя файлы с рейтингами фильмов из прошлого домашнего занятия (файл movies.csv и ratings.csv из базы https://grouplens.org/datasets/movielens). Т. е. проверить верно ли, что с ростом года выпуска фильма его средний рейтинг становится ниже.

```python
years = list(range(1950, 2011))

def production_year(param):
    """Определение года выпуска фильма"""
    for year in years:
        if str(year) in param:
            return year
    return 1900
```

```python
movies = pd.read_csv('ml-latest-small/movies.csv')
ratings = pd.read_csv('ml-latest-small/ratings.csv')

# Добавление столбца с информацией о годе производства фильма
movies['year'] = movies['title'].apply(production_year)

# Подсчет среднего рейтинга по каждому фильму
rates = ratings.groupby('movieId').mean()

# Объединение данных о рейтинге и названии фильма по столбцу movieId
union_rates_titles = pd.merge(movies, rates, on = 'movieId', how = 'inner')

# Определение среднего рейтинга фильмов по году производства
rate_by_year = union_rates_titles.groupby('year')['rating'].mean()

# Сортировка рейтинга по убыванию
rate_by_year.sort_values(ascending = False).head(20)
```
