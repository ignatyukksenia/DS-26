### Задание 1
Скачать с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определить какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd
```

```python
ratings = pd.read_csv('ratings.csv')
movies = pd.read_csv('movies.csv')

# Отсев фильмов с оценкой меньше 5.0, подсчет количества оценок по каждому фильму
max_rates = ratings[ratings.rating == 5.0].groupby(['movieId']).count()

# Объединение данных по столбцу movieId
union_rates_titles = pd.merge(max_rates, movies, on = 'movieId', how = 'inner')

# Сортировка по столбцу rating
top_rates= union_rates_titles.sort_values('rating', ascending = False)         
top_rates.head()
```

### Задание 2. По данным файла power.csv посчитать суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывать в расчетах отрицательные значения quantity.

```python
file = pd.read_csv('power.csv')

# Фильстрация по стране, категории, году и количеству потребленной электроэнергии
filter_by_country = file[(file.country == 'Estonia')|(file.country == 'Latvia')|(file.country == 'Lithuania')]
filter_by_category = filter_by_country[(filter_by_country.category == 4)|(filter_by_country.category == 12)|(filter_by_country.category == 21)]
filter_by_year = filter_by_category[(filter_by_category.year >= 2006)& (filter_by_category.year <= 2011)]
filter_by_quantity = filter_by_year[filter_by_year.quantity > 0]

# Подсчет общего потребления потребления по трем странам
total_consumption = filter_by_quantity.quantity.sum()

total_consumption
```
