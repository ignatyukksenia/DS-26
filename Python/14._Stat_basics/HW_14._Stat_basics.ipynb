## Задание 2. Работа с выбросами
В выбранных числовых столбцах найти выбросы, выдвинуть гипотезы об их причинах и проинтерпретировать результаты. Принять и обосновать решение о дальнейшей работе с ними.

```python
# Определение выбросов в столбцах с количественными данными

for column in quantitative_columns:
    print(column)
    # Расчет межквартильного размаха 
    Q1 = table[column].quantile(0.25)
    Q3 = table[column].quantile(0.75)
    IQR = Q3 - Q1
    
    # Расчет нижнего и врхнего значений выбросов
    lower_outlier = Q1 - (1.5 * IQR)
    higher_outlier = Q3 + (1.5 * IQR)
    print(f'lower_outlier: {lower_outlier}, higher_outlier: {higher_outlier}')
    remove_outliers = table[table[column].between(lower_outlier, higher_outlier, inclusive=True)].sort_values(column)
    outlier = pd.concat([table, remove_outliers]).drop_duplicates(keep=False)
    print(outlier[column].unique())
    print()
```

## Вывод:

### Пульс(pulse):
Сердечно-сосудистая система лошади совместима с диапазоном частоты сердечных сокращений от 20 до 240 ударов в минуту. Поэтому имеющие выбросы можно использовать при дальнейшей работе. Возможно, повышенные показатели пульса следует выделить в отдельную группу.

### Дыхательные движения(respiratory rate):
Частота дыхательных движений в норме составляет 8-10, при болезни и повышении нагрузки увеличивается (по некоторым данным под нагрузкой до 40-60 движений в минуту). Кроме того, в сопроводительной файле указано, что полезность информации сомнительна из-за больших колебаний. В это случае значения, превышающие расчетное значение не будут учитываться в дальнейшей работе.


### Объем упакованных клеток(packed cell volume):
Резерв эритроцитов селезенки лошади способен удвоить объем упакованных клеток во время максимальной нагрузки. Если исходить из этого утверждения, выбросы можно использовать в дальнейшемб т.к. значения не намного превышают рассчитанное верхнее значение. 

### Общий белок (total protein):
В столбце нет выбросов, только пропуски значений. 

## Задание 3. Работа с пропусками¶
Рассчитать количество пропусков для всех выбранных столбцов. Принять и обосновать решение о методе работы с пропусками по каждому столбцу, сформировать датафрейм, в котором пропуски будут отсутствовать.

```python
# Определение количества пропусков

for column in table.columns:
    missings = table[column].isna().mean()
    print(f'{column} - {missings:.1%}')
```

### Выводы:

'surgery' и 'outcome': В каждом столбце прошущено по одному значения, причем в одной и той же строке. В этом случае строка удаляется полностью, т.к. восстановить значения не представляется возможным.

'age': пропусков нет.

'peripheral pulse': пропущено 23% значений. Целесообразно заменить пропуски константой и в дальнейшем анализировать как отдельную категорию.

'pulse':
"концентрация" значений между первым и вторым квартилями больше, чем в остальных. Пропуски заполним модальным занчением, т.к. к указанному диапазону ближе мода.

'respiratory rate' Основная часть показателей находится между нижним и верхним квартилями. Мода и среднее знаечние приблизительно одинаково отстоят от медианы. При этом в диапазоне 28.0 - 40.0 показатели расположены наиболее плотно. К указанному диапазону ближе всего среднее значение, поэтому пропуски будут заполнени именно им.

'packed cell volume': Значения сгруппированы между нижним и верхним квартилями. При этом среднее значение близко к медиане, модальное - значительно отстоит. Поэтому пропуски будут заполнены средним значением.

'total protein': большая часть значений находится до второй квартили. Ко второй квартиле ближе модальное значение, пропуски будут заполнены модой.

```python
# Распределение показателей в столбце 'respiratory rate'
table.groupby('respiratory rate').count()
```

```python
# Удаление строки с пустым значением 'surgery' и 'outcome'
table = table[table['surgery'].notnull()]

# Заполние пропусков в столбцах

table['peripheral pulse'].fillna('constant', inplace = True)
table['pulse'].fillna(table['pulse'].mode()[0], inplace = True)
table['respiratory rate'].fillna(table['respiratory rate'].mean(), inplace = True)
table['packed cell volume'].fillna(table['packed cell volume'].mean(), inplace = True)
table['total protein'].fillna(table['total protein'].mode()[0], inplace = True)

table
```
